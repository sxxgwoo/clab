```python
import pandas as pd
from datetime import datetime
import os
day_offsets = [i * 24 * 60 for i in range(0, 8)]

# 시간과 분을 분 단위로 변환하는 함수
def convert_to_minutes(time_string, days_in_minutes):
    dt = datetime.strptime(time_string, '%Y-%m-%dT%H:%M:%S%z')
    hours = dt.hour
    minutes = dt.minute

    total_minutes = hours*60 + minutes + days_in_minutes
    return total_minutes

def indexing(i,dt):
    
    if i == 6 & dt.day == 29:
        index= 7
    else:
        index=dt.day - 21
        
    return index

# 주어진 조건을 만족하는 행을 필터링하는 함수
def filter_rows(df, i):
    # start_time, target_time에서 날짜 정보 추출
    df['start_time_dt'] = pd.to_datetime(df['start_time'], format='%Y-%m-%dT%H:%M:%S%z')
    df['target_time_dt'] = pd.to_datetime(df['target_time'], format='%Y-%m-%dT%H:%M:%S%z')
    df['req_time_dt'] = pd.to_datetime(df['req_leaving_time_timemachine'], format='%Y-%m-%dT%H:%M:%S%z')
    

    # days_in_minutes 계산을 위한 벡터화된 함수를 사용하여 start_time과 target_time에 적용
    df['days_in_minutes_start'] = df['start_time_dt'].apply(lambda x: day_offsets[indexing(i,x)])
    df['days_in_minutes_target'] = df['target_time_dt'].apply(lambda x: day_offsets[indexing(i,x)])
    df['days_in_minutes_req'] = df['req_time_dt'].apply(lambda x: day_offsets[indexing(i,x)])

    # cur_time_minutes, path_time_LastAPI_minutes, target_time_minutes 계산
    df['cur_time_minutes'] = df.apply(lambda row: convert_to_minutes(row['cur_time'], row['days_in_minutes_start']), axis=1)
    df['path_time_LastAPI_minutes'] = df['path_time_LastAPI'] / 60
    df['target_time_minutes'] = df.apply(lambda row: convert_to_minutes(row['target_time'], row['days_in_minutes_target']), axis=1)
    df['req_leaving_time_timemachine'] = df.apply(lambda row: convert_to_minutes(row['req_leaving_time_timemachine'], row['days_in_minutes_req']), axis=1)
    

    # 주어진 조건 abs((cur_time + path_time_LastAPI) - (target_time - 8)) <= 1
    df_filtered = df[abs((df['cur_time_minutes'] + df['path_time_LastAPI_minutes']) - (df['target_time_minutes'] - 8)) < 1]

    # 같은 id에 대해 중복된 행 제거
    df_unique = df_filtered.drop_duplicates(subset='id')

    # 필요 없는 중간 계산 열 삭제 (옵션)
    df_unique = df_unique.drop(columns=['start_time_dt', 'target_time_dt', 'days_in_minutes_start', 'days_in_minutes_target'
                                        ,'path_TimeMachine','sub_path_time_LastAPI','sub_path_time_TimeMachine','path_time_ans','req_time_dt','days_in_minutes_req','path_LastAPI'])

    return df_unique

# 처리할 파일 목록
base_path = '/home/sxxgwoo/clab/data/real'
file_names = [f'2024092{i}.csv' for i in range(2, 9)]

# 각 파일에 대한 분 단위 조정값 (21일은 6*24*60, 23일은 24*60 등)


# print(day_offsets)
# 파일 처리 및 저장
for i, file_name in enumerate(file_names):
    file_path = os.path.join(base_path, file_name)

    # 파일 읽기
    df = pd.read_csv(file_path)

    # 해당 날짜에 맞는 분 단위 조정값 사용
    timemachine_df = df[df['group'] == 'timemachine'][['id', 'req_leaving_time']]

    # 'id'를 기준으로 원본 df에 새로운 열 'timemachine' 추가 (같은 'id'에 대해 매핑)
    df = pd.merge(df, timemachine_df[['id', 'req_leaving_time']], on='id', how='left', suffixes=('', '_timemachine'))

    # 필터링된 DataFrame 추출
    print(i)
    filtered_df = filter_rows(df, i)
    num_rows = filtered_df.shape[0]
    
    # 결과 파일 저장
    output_file_name = f'filtered_rows_{file_name[-6:-4]}.csv'  # 날짜를 파일명에 포함
    output_path = os.path.join(base_path, 'filtered', output_file_name)
    filtered_df.to_csv(output_path, index=False)

    print(f"Processed and saved: {output_file_name} with {num_rows} rows")
```

```python
import pandas as pd
import os

# 처리할 파일 목록
base_path = '/home/sxxgwoo/clab/data/real/filtered'
file_names = [f'filtered_rows_2{i}.csv' for i in range(2, 9)]

# 빈 리스트를 만들어서 각 파일에서 랜덤 100개의 데이터를 담을 예정
df_list = []

# 각 파일에서 랜덤 100개의 데이터 추출
for file_name in file_names:
    file_path = os.path.join(base_path, file_name)
    
    # 파일 읽기
    df = pd.read_csv(file_path)
    
    # 랜덤으로 100개 추출 (replace=False는 중복 추출 방지)
    random_100 = df.sample(n=100, replace=False, random_state=42)
    
    # 리스트에 추가
    df_list.append(random_100)

# 모든 DataFrame을 하나로 합치기
combined_df = pd.concat(df_list, ignore_index=True)

# 합친 DataFrame을 새로운 CSV 파일로 저장
output_file_path = '/home/sxxgwoo/clab/data/real/filtered/combined_random_700.csv'
combined_df.to_csv(output_file_path, index=False)

# 완료 메시지 출력
print(f"Combined random 100 rows from each file saved as: {output_file_path}")
```

```python
import pandas as pd
import numpy as np
import os
import math
from datetime import datetime

file_path = '/home/sxxgwoo/clab/data/real/filtered/test_set.csv'
df=pd.read_csv(file_path)


time_interval=1
episode =0
df_final = pd.DataFrame()
# df_test=df.head(2)
# df_test
for i , row in df.iterrows():
    
    if row['cur_time_minutes'] > row['req_leaving_time_timemachine']:
        repeat = 61 + math.ceil((row['cur_time_minutes'] - row['req_leaving_time_timemachine'])/time_interval)
    else:
        repeat = 61
    # print(repeat)
    # print(row['req_leaving_time_timemachine'])
    # print(row['cur_time_minutes'])
    row['start_time'] = row['req_leaving_time_timemachine'] - 60
    row['target_time'] = row['target_time_minutes'] -8


    row_df =  pd.Series(row).to_frame().T
    # row_df

    rows_repeat_df = pd.concat([row_df]*int(repeat), ignore_index=True)
    rows_repeat_df

    rows_repeat_df['cur_time'] = time_interval
    rows_repeat_df['cur_time'][0] = row['req_leaving_time_timemachine'] - 60
    rows_repeat_df['cur_time'] = rows_repeat_df['cur_time'].cumsum()
    rows_repeat_df['remain_time_from_TimeMachine'] = rows_repeat_df['req_leaving_time_timemachine'] - rows_repeat_df['cur_time']
    rows_repeat_df['oracle']= row['cur_time_minutes']

    rows_repeat_df['round'] = 1
    rows_repeat_df['round'] = rows_repeat_df['round'].cumsum() -1
    rows_repeat_df['episode'] = episode
    episode +=1


    df_final = pd.concat([df_final ,rows_repeat_df], ignore_index=True)
df_final
save_path='/home/sxxgwoo/clab/data/real/filtered/test_set_offline.csv'
os.makedirs(os.path.dirname(save_path), exist_ok=True)
df_final.to_csv(save_path, index=False)

print("file saved successfully")
```
